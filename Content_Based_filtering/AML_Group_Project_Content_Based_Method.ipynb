# Content-Based Filtering

## **Feature selelction and preprocession**

```python
transactions_sample = pd.read_csv('m1_train.csv',converters={'article_id': str})
articles_sample = pd.read_csv('articles_sample.csv',converters={'article_id': str})
```

```python
transactions_sample['article_id'] = '0' + transactions_sample['article_id']
```

```python
# Join transaction and article dataframe
df = transactions_sample.merge(articles_sample, on='article_id')

# Feature selection
df = df[['t_dat', 'customer_id', 'article_id', 'prod_name', 'product_type_name',
       'product_group_name', 
       'graphical_appearance_name', 'colour_group_name',
       'perceived_colour_value_name',
       'perceived_colour_master_name',
       'department_name', 'index_name',
       'index_group_name', 'section_name',
       'garment_group_name', 'detail_desc']]
feature_subset = ['product_group_name', 
       'graphical_appearance_name', 'colour_group_name',
       'perceived_colour_value_name',
       'perceived_colour_master_name',
       'department_name', 'index_name',
       'index_group_name', 'section_name',
       'garment_group_name']

# One hot encoding
features = feature_subset
df1 = df[['customer_id', 'article_id'] + features]
dummies_df = pd.get_dummies(df1, columns=features)
dummies_df
```

```python
# select users who bought at least 2 items and preprocessing
minimum_items = 0
groupby_customer = dummies_df.groupby('customer_id')

feature_score = []  
cutomer_ids = []
article_ids = []
for key in groupby_customer.groups.keys():
    temp = groupby_customer.get_group(key) # get sub_dataframe for each customer transactions
    if temp.article_id.nunique() >= minimum_items:
        # feature_score is the sum of # of transaction for each feature
        feature_score.append(temp.drop('article_id', axis=1).sum(numeric_only=True).values)
        cutomer_ids.append(key)
        article_ids.extend(temp.article_id.values.tolist())
```

## **Create user feature matrix**

Check [here](https://www.youtube.com/watch?v=C7m2IWUvkys&list=PLTjbQZu8nnshv8H5tV0fGh0ECBIXwuHJb&index=9) for user feature matrix explaination

```python
user_feature = pd.DataFrame(feature_score, columns = dummies_df.columns[2:])
user_sum_socre = user_feature.sum(axis=1)
normalized_user_feature = user_feature.div(user_sum_socre, axis=0)
normalized_user_feature.insert(0, 'customer_id', cutomer_ids)
normalized_user_feature = normalized_user_feature.set_index('customer_id')
normalized_user_feature
```

## **Create item feature  matrix** 

```python
unique_items = dummies_df.drop_duplicates(subset='article_id')

# filter out the items been bought by customers
bought_item = unique_items.article_id.isin(article_ids)

item_feature = unique_items[bought_item].drop('customer_id', axis=1)

item_feature = item_feature.set_index('article_id')
item_feature
```

## Calculate customer interest score

Check [here](https://www.youtube.com/watch?v=HjGTxoYajow&list=PLTjbQZu8nnshv8H5tV0fGh0ECBIXwuHJb&index=10) at 0:28 for formula explaination

```python
scores = normalized_user_feature.dot(item_feature.T)
scores
```

## Get recommendations

```python
def get_rcmnd(customer_id, scores):
    cutomer_scores = scores.loc[customer_id]
    customer_prev_items = groupby_customer.get_group(customer_id)['article_id']
    prev_dropped = cutomer_scores.drop(customer_prev_items.values)
    ordered = prev_dropped.sort_values(ascending=False)   
    return ordered[:20]
```

```python
# prediction: build recommendation for each customer

# filtered_customer_ids: customer ids with >=2 transactions
filtered_customer_ids = transactions_sample.groupby(['customer_id'])['customer_id'].count()
filtered_customer_ids = filtered_customer_ids[filtered_customer_ids >= 2].index.tolist()

customer_to_prediction_dict = {}
counter = 0
for id in list(test_cust_df['customer_id']):
  if counter % 1000 == 0:
    print('progress:', counter / len(filtered_customer_ids), '%')
  counter += 1
  t = get_rcmnd(id,scores)
  t = list(t.index)
  if t is not None:
    customer_to_prediction_dict[id] = t
```

```python
joined_set = pd.DataFrame(customer_to_prediction_dict.keys(),columns=['customer_id'])
joined_set['recc_buy']=0
for index, n in enumerate(customer_to_prediction_dict.values()):
    joined_set['recc_buy'][index] = n
joined_set
```

```python
truth_df = pd.merge(test_cust_df, joined_set, left_on='customer_id', right_on='customer_id', how='left')
truth_df
```

```python
def apk2(actual, predicted):
  num_hits = 0.0

  for i,p in enumerate(actual):
    if p in predicted:
      num_hits += 1.0

  return num_hits / len(actual)

def mapk2(df):
  s = []
  for idx, row in df.iterrows():
    s.append(apk2(row['actual_buy'], row['recc_buy']))
  return np.mean(s)

print(f'Final MAP: {mapk2(truth_df)}')
```
